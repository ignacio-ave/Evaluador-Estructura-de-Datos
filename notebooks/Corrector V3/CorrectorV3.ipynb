### Parte 1 : Cargar datos 

```python
import json
import pandas as pd
import re

# 1. Solicitar la ruta y el nombre del archivo JSON al usuario
file_path = input("Por favor, ingresa la ruta completa del archivo JSON (incluyendo el nombre del archivo): ")

# 2. Cargar el archivo JSON
with open(file_path, "r") as file:
    data = json.load(file)

# 3. Funciones auxiliares para procesar el archivo JSON
def flatten_list(l):
    """Función para aplanar listas anidadas"""
    flat_list = []
    for item in l:
        if isinstance(item, list):
            flat_list.extend(flatten_list(item))
        else:
            flat_list.append(item)
    return flat_list

def extract_from_json_adjusted(data):
    extracted_data = []
    for student_data in data:
        flattened_data = flatten_list(student_data)
        email_pattern = r"\S+@\S+"
        date_pattern = r"\d{1,2} de \w+ de \d{4}"
        time_pattern = r"\d+ minutos \d+ segundos"
        email = re.search(email_pattern, ' '.join(flattened_data)).group()
        dates = re.findall(date_pattern, ' '.join(flattened_data))
        start_date = dates[0] if len(dates) > 0 else None
        end_date = dates[1] if len(dates) > 1 else None
        time_spent = re.search(time_pattern, ' '.join(flattened_data)).group() if re.search(time_pattern, ' '.join(flattened_data)) else None
        name = flattened_data[0]
        status = flattened_data[3]
        answers = flattened_data[8:]
        student_info = [name, email, status, start_date, end_date, time_spent] + answers
        extracted_data.append(student_info)
    return extracted_data

# 4. Procesar los datos
students_data = data[0]
all_students_data_adjusted = extract_from_json_adjusted(students_data)
max_columns = max(len(student) for student in all_students_data_adjusted)

# Misma cantidad de columnas para todos los estudiantes
for student in all_students_data_adjusted:
    while len(student) < max_columns:
        student.append(None)

# Generar el DataFrame
num_questions = (max_columns - 6) // 2
adjusted_columns = ["Nombre", "Email", "Estado", "Inicio", "Fin", "Tiempo"] + [f"Q{i//2+1}" if i % 2 == 0 else f"A{i//2+1}" for i in range(num_questions * 2)]
df_all_students_final_adjusted = pd.DataFrame(all_students_data_adjusted, columns=adjusted_columns)

# 5. Guardar en Excel y mostrar la ubicación del archivo guardado
output_path = input("Por favor, ingresa la ruta donde deseas guardar el archivo Excel (sin incluir el nombre del archivo): ")
output_file_name = input("Por favor, ingresa el nombre que deseas para el archivo Excel (por ejemplo, 'mi_archivo.xlsx'): ")
df_all_students_final_adjusted.to_excel(output_path + '/' + output_file_name, index=False)
print(f"¡El archivo ha sido guardado en {output_path}/{output_file_name}!")
```

```python
import hashlib

def reorganize_dataframe(df):
    # Creamos una lista para almacenar los datos reorganizados
    data = []
    # Recorremos cada fila del DataFrame original
    for index, row in df.iterrows():
        # Recorremos cada pregunta
        for i in range(1, 13):  # Asumimos que hay 12 preguntas (Q1, Q2, ..., Q12)
            # Añadimos los datos reorganizados a la lista
            data.append({
                'Nombre': row['Nombre'],
                'Pregunta': row[f'Q{i}'],
                'Respuesta': row[f'A{i}']
            })
    # Creamos un nuevo DataFrame con los datos reorganizados
    new_df = pd.DataFrame(data)
    # Añadimos el ID hash
    new_df['ID'] = new_df['Pregunta'].apply(lambda x: hashlib.sha256(x.encode()).hexdigest())
    # Configuramos el multiíndice
    new_df.set_index(['Nombre', 'Pregunta'], inplace=True)
    return new_df

new_df = reorganize_dataframe(df)
```

```python
from pydantic import BaseModel, Field, validator
import pandas as pd
from typing import List


class EvaluationCriteria(BaseModel):
    razonamiento: str = Field(description="Razonamiento sobre el criterio")
    puntaje: float = Field(description="Puntaje asignado")

    @validator('puntaje')
    def check_score(cls, puntaje):
        if puntaje < 0 or puntaje > 3:
            raise ValueError("La puntuación debe estar entre 0 y 3")
        return puntaje

class EvaluationJSON(BaseModel):
    estructura: str
    pregunta: str
    respuesta: str
    correctitud: EvaluationCriteria
    claridad: EvaluationCriteria
    relevancia: EvaluationCriteria

    def to_dict(self):
        return {
            "estructura": self.estructura,
            "pregunta": self.pregunta,
            "respuesta": self.respuesta,
            "correctitud": self.correctitud.dict(),
            "claridad": self.claridad.dict(),
            "relevancia": self.relevancia.dict()
        }

class QuestionInfo:
    def __init__(self, question_id, student, topic, question, total_points, criteria, scoring_guideline, good_response, bad_response):
        self.question_id = question_id  # ID hash
        self.student = student
        self.topic = topic
        self.question = question
        self.total_points = total_points
        self.criteria = criteria
        self.scoring_guideline = scoring_guideline
        self.good_response = good_response
        self.bad_response = bad_response
        self.evals = []

    def to_dict(self):
        return {
            "student": self.student,
            "estructura": self.topic,
            "pregunta": self.question,
            "respuesta": self.good_response,
            "correctitud": {
                "razonamiento": None,
                "puntaje": None
            },
            "claridad": {
                "razonamiento": None,
                "puntaje": None
            },
            "relevancia": {
                "razonamiento": None,
                "puntaje": None
            }
        }

```

```python

# Leer "pauta.xlsx" 
import pandas as pd 

def load_questions_from_excel(filepath):
    pauta_df = pd.read_excel(filepath)
    questions = []
    for index, row in pauta_df.iterrows():
        question_info = QuestionInfo(
            question_id=row["ID"],
            student="Sample Student",  # Ejemplo para la desmostracion de la pauta
            topic=row["Tópico"],
            question=row["Pregunta"],
            total_points=row["Total Puntos"],
            criteria=row["Scoring Guideline"],
            scoring_guideline=row["Scoring Guideline"],
            good_response=row["Response GPT-4"],
            bad_response=row["Bad Response GPT-4"]
        )
        questions.append(question_info)
    return questions

questions = load_questions_from_excel(input("Ingrese el nombre del archivo y su ruta: "))

for index, question in enumerate(questions, start=1):
    print("-------------------------")
    print(f"ID: {question.question_id}")
    print(f"Pregunta {index}:")
    print(f"Estudiante: {question.student}")
    print(f"Tópico: {question.topic}")
    print(f"Pregunta: {question.question}")
    print(f"Puntos totales: {question.total_points}")
    print(f"Criterio de evaluación: {question.criteria}")
    print(f"Pauta de evaluación: {question.scoring_guideline}")
    print(f"Respuesta correcta: {question.good_response}")
    print(f"Respuesta incorrecta: {question.bad_response}")
```

```python
def generate_prompts_from_dataframe(df, questions, jsonformat):
    # Creamos un diccionario para mapear las preguntas a sus objetos correspondientes
    question_dict = {question.question: question for question in questions}

    prompts = []
    # Recorremos cada fila del DataFrame
    for (student, question_text), row in df.iterrows():
        # Buscamos la información de la pregunta correspondiente
        question = question_dict.get(question_text)

        # Si no encontramos la información de la pregunta, utilizamos valores por defecto
        if question is None:
            topic = "Desconocido"
            scoring_guideline = "Desconocido"
            criteria = "Desconocido"
        else:
            topic = question.topic
            scoring_guideline = question.scoring_guideline
            criteria = question.criteria

        # Generamos el prompt
        prompt = f'''
        Actúa como un evaluador universitario de la materia ESTRUCTURA DE DATOS. En esta materia se evalúa el conocimiento de los estudiantes del departamento de informática en estructura de datos bien conocida.
        Las respuestas serán teóricas y breves.

        La pregunta a evaluar trata sobre la estructura {topic}
        Información de la Estructura: {scoring_guideline}

        Pregunta: {question_text}
        Respuesta Proporcionada: {row['Respuesta']}

        Criterios de evaluación: {criteria}

        Por favor, sigue los siguientes pasos para evaluar la respuesta según los criterios de evaluación:
        1. Analiza la estructura y la información proporcionada.
        2. Revisa la pregunta y la respuesta proporcionada.
        3. Razone sobre la adecuación de la respuesta a la pregunta, considerando los criterios específicos para la estructura en cuestión.
            - Razonamiento sobre la Correctitud de la Respuesta
            - Razonamiento sobre la Claridad de la Respuesta
            - Razonamiento sobre la Relevancia de la Respuesta
        4. Rellena el archivo JSON de evaluación adjunto, incluyendo tus razonamientos y el puntaje asignado en cada criterio.
        5. Devuelve el archivo JSON completado.
        
        Formato del JSON:
        {jsonformat}
        '''
        prompts.append(prompt)

    return prompts
```

```python
import os
import openai
from getpass import getpass

def get_openai_key_from_input():
    """Obtiene el token del API de OpenAI a través de una entrada segura"""
    try:
        openai_api_key = getpass('Por favor, ingrese su token de API de OpenAI: ')
        if not openai_api_key:
            raise ValueError('No se proporcionó el token de API de OpenAI.')
        return openai_api_key
    except Exception as e:
        print(f'Error obteniendo el token de API de OpenAI: {e}')
        raise

def init_openai():
    """Inicializa el cliente de OpenAI con el token del API"""
    try:
        openai.api_key = get_openai_key_from_input()
    except Exception as e:
        print(f'Error inicializando OpenAI: {e}')
        raise

init_openai()
```

```python
def generate_responses(prompts, max_tokens=500):
    responses = []
    with open("output.txt", "w") as f:
        for i, prompt in enumerate(prompts):
            messages = [{'role': 'system', 'content': prompt}]
            try:
                response = openai.ChatCompletion.create(
                    model="gpt-3.5-turbo",
                    messages=messages,
                    max_tokens=max_tokens
                )
                response_content = response.choices[0].message['content']
                responses.append(response_content)

                # Escribir en el archivo
                f.write(f"###Human:{prompt}\n###Assistant:{response_content}\n\n")

                # Imprimir la respuesta
                print(f"Response {i+1}:\n{response_content}\n{'-'*50}")

            except Exception as e:
                print(f"Error occurred: {e}")
                responses.append(None)
    return responses

responses = generate_responses(prompts)
```
